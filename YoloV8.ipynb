#Train vá»›i custom data


```python
#Giáº£ nÃ©n file data vÃ  tá»• chá»©c vÃ o thÆ° má»¥c train

#Extract zip data file
%cd /content/drive/MyDrive/Dataset
!unzip /content/drive/MyDrive/Dataset/Sign-Language-Digits-Dataset-master.zip

#Tá»• chá»©c táº¡i cÃ¡c file
%cd /content/drive/MyDrive/Dataset
!mkdir train
!mkdir train/images
!mkdir train/labels
!mv *.jpg train/images
!mv *.txt train/labels
!unzip -o /content/drive/MyDrive/Dataset/Sign-Language-Digits-Dataset-master.zip
```

```python
#Create training yaml file
%cd /content/drive/MyDrive/Dataset

!rm -f mydataset.yaml
!echo "train: /content/drive/MyDrive/Dataset/train" >> mydataset.yaml
!echo "val: /content/drive/MyDrive/Dataset/val" >> mydataset.yaml
!echo "nc: 1" >> mydataset.yaml
!echo "names: ['fire']" >> mydataset.yaml
```

```python
%%bash
set -e

base="/content/drive/MyDrive/Dataset/Sign-Language-Digits-Dataset-master/Dataset"
cd "$base"

mkdir -p train val

for i in {0..9}; do
  mkdir -p "train/$i" "val/$i"

  # láº¥y danh sÃ¡ch file áº£nh trong lá»›p i (Ä‘Ãºng vá»›i .JPG/.jpg/.png)
  mapfile -t files < <(ls -1 "$i" | shuf)

  n=${#files[@]}      # âœ… sá»‘ lÆ°á»£ng file
  val_n=$((n/5))      # 20%

  for ((j=0; j<n; j++)); do
    f="${files[j]}"
    if (( j < val_n )); then
      mv "$i/$f" "val/$i/"
    else
      mv "$i/$f" "train/$i/"
    fi
  done
done

echo "Done split dataset!"
```

```python
#Train vá»›i CLI
!yolo task=classify mode=train model=yolov8n-cls.pt \
data=/content/drive/MyDrive/Dataset/Sign-Language-Digits-Dataset-master/Dataset \
epochs=10 imgsz=224
```
